```python
import numpy as np
from matplotlib import pyplot as plt
% pylab inline
```

```python
x = np.sqrt(3)


for N in range(1,15):
  rs = np.asarray([k*x - np.floor(k*x) for k in range(1, N+1)])
  ys = N*np.ones((len(rs)))

  plt.plot(rs, ys, 'x')
  eps = 1e-09
  reg = np.arange(1.0/(N+1),1-eps,1.0/(N+1))
#   plt.plot(reg, ys, 'o')

  ix = np.where(np.abs(rs - np.round(rs)) < 1.0/(N+1.0))[0]
  plt.plot(rs[ix], ys[ix], 'o')
  print(ix)
  
  


# plt.xticks( [ x/(N+1.0) for x in range(0, N+1)])
plt.xlim([0,1])
plt.ylim([-1, N+1])
plt.grid(True)

```

```python
N=200
x =  np.pi 
ns = np.arange(1,N+1)
xns = x*ns
e = 1.0 / (N+1.0)
ix = np.where(np.abs(xns - np.round(xns)) < e)
print(ns[ix])

plt.plot(ns, xns, 'x')
a = ns[ix]
b = xns[ix]
plt.plot(a, b , 'o')
plt.plot(a, np.floor(b), '.')
plt.plot(a, np.ceil(b), '.')
plt.xlim([0,N+1])
plt.grid(True)
```

```python
for b in [1.0, 7.0, 113]:
  print('\n  b= {:}\n -----------'.format(b))
  diff_down = b*x - np.floor(b*x)
  print('bx - floor bx {}'.format(diff_down ))

  diff_up = np.ceil(b*x) - b*x
  print('ceil bx - bx  {}'.format(diff_up))
  if diff_down < diff_up:
    a = np.floor(b*x)
    print('floor bx: {}'.format(a))
  else:
    a = np.ceil(b*x)
    print('ceil bx: {}'.format(a))

  est = a / b
  print('est {}'.format(est))

  err = np.abs(est - x)
  print('err {:}'.format(err))

  print('1/b^2 {}'.format(1.0/b**2))
```
